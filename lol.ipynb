```python
from bs4 import BeautifulSoup
```

```python
tesla_ticker = yf.Ticker('TSLA')
tesla_data = tesla_ticker.history(period='max')
tesla_data_reset = tesla_data.reset_index()
tesla_data_head = tesla_data_reset.head()
print(tesla_data_head)

```

```python
gamestop_ticker = yf.Ticker('GME')
gamestop_data = gamestop_ticker.history(period='max')
gamestop_data_reset = gamestop_data.reset_index()
gamestop_data_head = gamestop_data_reset.head()
print(gamestop_data_head)
```

```python
url ='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm'
response = requests.get(url)
soup = BeautifulSoup(response.text,'html.parser')
table = soup.find_all('table')[1]
rows = table.find_all('tr')
tesla_revenue = []
for rows in rows[1:]:
    cols = rows.find_all('td')
    date = cols[0].text.strip()
    revenue = cols[1].text.strip()
    tesla_revenue.append((date,revenue))
tesla_ravenue_df = pd.DataFrame(tesla_revenue,columns=['date','revenue'])
print(tesla_ravenue_df.tail())
```

```python
def make_graph (stock_data, tiitle):
    plt.figure(figsize=(29,8))
    plt.plot(stock_data.index, stock_data['Close'], label= 'Stock Price')
    plt.title('Tesla Stock Data')
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.legend()
    plt.grid(True)
    plt.show()

make_graph(tesla_data, 'Tesla')
```
